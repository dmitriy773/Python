```python
#10
price = float(input("Цена за 1 кг: "))

weight = 1.0
while weight <= 2.01:  # 2.01 для учета погрешности float
    cost = price * weight
    print(f"{weight:.1f} кг - {cost:.2f} руб.")
    weight += 0.2
```

```python
#14
N = int(input("N: "))

K = 0
while 3 ** (K + 1) < N:
    K += 1

print("Наибольшее K:", K)
```

```python
#15
N = int(input("N: "))
A = float(input("A: "))
B = float(input("B: "))

H = (B - A) / N
print("Длина отрезка H:", H)

current = A
for i in range(N + 1):
    print(f"{current:.2f}")
    current += H
```
